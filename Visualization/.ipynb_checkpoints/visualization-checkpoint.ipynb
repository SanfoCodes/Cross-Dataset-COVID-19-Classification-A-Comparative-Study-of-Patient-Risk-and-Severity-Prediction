```python
# Visualization notebook for comparing Modeling 1 & 2 results across all 5 datasets

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
sns.set(style="whitegrid")
```

```python
# === Plot 1: Accuracy Comparison ===
plt.figure(figsize=(12, 6))
sns.barplot(data=results_df, x="Dataset", y="Accuracy", hue="Model")
plt.title("Model Accuracy Comparison Across Datasets")
plt.ylim(0.7, 1.0)
plt.legend(title="Model")
plt.show()

# === Plot 2: F1 Score Comparison ===
plt.figure(figsize=(12, 6))
sns.barplot(data=results_df, x="Dataset", y="F1_Score", hue="Model")
plt.title("Model F1 Score Comparison Across Datasets")
plt.ylim(0.7, 1.0)
plt.legend(title="Model")
plt.show()
```

```python
# === Plot 3: Precision Comparison ===
plt.figure(figsize=(12, 6))
sns.lineplot(data=results_df, x="Dataset", y="Precision", hue="Model", marker='o')
plt.title("Model Precision Trends")
plt.ylim(0.7, 1.0)
plt.show()

# === Plot 4: Recall Comparison ===
plt.figure(figsize=(12, 6))
sns.lineplot(data=results_df, x="Dataset", y="Recall", hue="Model", marker='o')
plt.title("Model Recall Trends")
plt.ylim(0.7, 1.0)
plt.show()
```

```python
# === Plot 5: Heatmap of Accuracy ===
heatmap_data = results_df.pivot(index="Dataset", columns="Model", values="Accuracy")
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title("Heatmap of Model Accuracy per Dataset")
plt.show()
```
